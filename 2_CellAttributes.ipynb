```python
### Packages
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
### Generate a random graph
### Assume we have processed the polygon file and constructed a graph network
num_nodes = 10  # Number of nodes
probability = 0.4  # Probability of edge creation
G = nx.binomial_graph(num_nodes, probability)
nx.draw(G, with_labels=True, node_size=700, cmap=plt.cm.get_cmap('viridis'))
```

```python
SavedAttributes = pd.read_csv("Syn_attributes.tsv", sep="\t", header=0, index_col=0)
# SavedAttributes = pd.read_csv(SynPath, sep="\t", header=0, index_col=0)
SavedAttributes
```

```python
### Add Node attributes
nx.set_node_attributes(G, SavedAttributes['Red'], "Biomarker1")
nx.set_node_attributes(G, SavedAttributes['Blue'], "Biomarker2")
nx.set_node_attributes(G, SavedAttributes['Green'], "Biomarker3")
nx.set_node_attributes(G, SavedAttributes['Colour'], "NodeColour")
```

```python
nx.draw(G, with_labels=True, node_size=700, cmap=plt.cm.get_cmap('viridis'))
```

```python
# Visualize the graph with node colors based on class
pos = nx.spring_layout(G)
node_colors = [G.nodes[i]['NodeColour'] for i in G.nodes()]
nx.draw(G, node_color=node_colors, with_labels=True, node_size=700, cmap=plt.cm.get_cmap('viridis'))

# # Show the plot
# plt.title("Random Graph Network with Node Classes")
plt.show()
```
