```python
### Loading Packages
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create an empty undirected graph
G = nx.Graph()
```

```python
# Add nodes
G.add_nodes_from(["Charla", "Connor", "Jia", "Murali", "Raeesah"])
```

```python
# Add edges (representing friendships)
G.add_edge("Charla", "Connor")
G.add_edge("Charla", "Jia")
G.add_edge("Connor", "Jia")
G.add_edge("Charla", "Murali")
G.add_edge("Connor", "Murali")
G.add_edge("Jia", "Raeesah")
G.add_edge("Charla", "Raeesah")
G.add_edge("Murali", "Raeesah")
```

```python
# Draw the network
nx.draw_networkx(G)
plt.show()
```

```python
# Draw the network
plt.figure(figsize=(8, 6)) # Adjust figure size for better visualization

nx.draw_networkx(G)

plt.title("Social Network Example")
plt.axis('off') # Hide axes
plt.show()
```

```python
# Draw the network
plt.figure(figsize=(8, 6)) # Adjust figure size for better visualization

nx.draw_networkx(G, with_labels=True, node_color='lightgreen', node_size=2000, font_size=10, font_weight='bold')

plt.title("Social Network Example")
plt.axis('off') # Hide axes
plt.show()
```
